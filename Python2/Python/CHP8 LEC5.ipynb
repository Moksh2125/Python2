# Simple Web Browser in Python

```python
import socket
mysock=socket.socket()
mysock.connect(("data.pr4e.org",80))
cmd='GET http://data.pr4e.org/clown.txt HTTP/1.0\r\n\r\n'.encode()
mysock.send(cmd)
while True:
    data=mysock.recv(512)
    if len(data)<1:
        break
    print(data.decode(),end="")
mysock.close()
```

# Simple HTTP Server in Python

```python
from socket import *
def createServer():
    serversocket=socket()
    serversocket.bind(("localhost",9000))
    serversocket.listen()
    while True:
        (clientsocket,address)=serversocket.accept()
        rd=clientsocket.recv(512).decode()
        print(rd)
        data="HTTP/1.1 200 OK\r\n"
        data+="Content-Type:text/html; charset=utf-8\r\n"
        data+="\r\n"
        data+="<html><body>Hello World</body></html>\r\n\r\n"
        clientsocket.send(data.encode())
    serversocket.close()
print("Access http://localhost:9000")
createServer()
```

# Politifact.com

```python
import requests
from bs4 import BeautifulSoup
website="https://www.politifact.com/factchecks/"
result=requests.get(website)
content=result.text
```

```python
soup = BeautifulSoup(content,'html.parser')
print(soup.prettify())
```

```python
facts = soup.findAll('li', class_='o-listicle__item')
```

```python
len(facts)
```

```python
# for names
names = []
for fact in facts:
    name = fact.find('div', class_='m-statement__meta').find('a').text.strip()
    names.append(name)
names
```

```python
# for statements
statements = []
for fact in facts:
    statement = fact.find('div', class_='m-statement__quote').text.strip()
    statements.append(statement)
statements
```

```python
# for dates
dates = []
for fact in facts:
    text = fact.find('footer', class_='m-statement__footer').text.strip()
    list = text.split('•')
    dates.append(list[1])
dates
```

```python
import pandas as pd
df = pd.DataFrame({'Source':names, 'Statement':statements, 'Date': dates})
df
```

# Offline

```python
soup = BeautifulSoup(open("top250.html","r",encoding="utf-8"))
print(soup.prettify())
```

```python
# for names
movie = soup.findAll('td', class_='titleColumn')
rating= soup.findAll('td',class_="ratingColumn imdbRating")
titles= []
dates=[]
scores=[]
for name in movie:
    title = name.find('a').text.strip()
    date=name.find('span').text.strip()[1:5]
    titles.append(title)
    dates.append(date)
for score in rating:
    ratings=score.find('strong').text.strip()
    scores.append(ratings)
print(titles)
print(dates)
print(scores)
```

```python
import pandas as pd
df = pd.DataFrame({'Movie':titles, 'Release Year':dates, 'Rating': scores})
df
```

# Subslikescript

```python
import requests
from bs4 import BeautifulSoup
website="https://subslikescript.com/movies"
result=requests.get(website)
content=result.text
soup = BeautifulSoup(content,'html.parser')
print(soup.prettify())
```

```python
box=soup.find_all('article',class_="main-article")
print(box)
```

```python
links=[]
for i in box:
    a=i.findAll('a',href=True)
    links.append(a)
```

```python
links=links[0]
```

```python
for i in links:
    print(i['href'])
```

```python
import requests
root = 'https://subslikescript.com' 
for link in links:
    website = root+link['href']
    result = requests.get(website)
    content = result.text
    soup = BeautifulSoup(content, 'html.parser')
    box = soup.find('article', class_='main-article')
    title = box.find("h1").text  # we can also use here get_text()
    transcript = box.find('div', class_='full-script').text.strip()
    try:
        with open(f"{title}.txt",'w', encoding='utf') as file:
            file.write(transcript)
            
    except Exception as e:
            print(e)
```
