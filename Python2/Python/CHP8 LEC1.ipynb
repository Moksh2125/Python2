# APIs:

```python
import requests, json
def get_coordinates():
    api_key=input("Enter API Key:")
    city=input("Enter City Name:")
    request_url="http://api.openweathermap.org/geo/1.0/direct?q="+city+"&appid="+api_key
    response=requests.get(request_url)
    coordinates=response.json()
    print(coordinates)
    Lat=coordinates[0]['lat']
    Lon=coordinates[0]['lon']
    print("Latitude:",Lat)
    print("Longitude:",Lon)
get_coordinates()
```

```python
import requests, json
def get_weather():
    api_key=input("Enter API Key:")
    city=input("Enter City Name:")
    url=f"https://api.openweathermap.org/data/2.5/weather?lat={23.0216238}&lon={72.5797068}&appid={api_key}"
    response=requests.get(url)
    coords=response.json()
    print(coords)
    print("Description:",coords['weather'][0]['description'])
    print("Temperature:",coords['main']['temp'])
    print("Pressure:",coords['main']['pressure'])
    print("Humidity:",coords['main']['humidity'])
    print("Wind Speed:",coords['wind']['speed'])
    print("Visibility:",coords['visibility'])
    
get_weather()
```
