## Dataframe
### Two dimensional data structure like 2D array or a table with rows and columns

```python
import pandas as pd
data = {'calories' : [400, 380, 420],
       'duration' : [50, 40, 45]}
df =  pd.DataFrame(data)
print(df)
```

```python
# Use to observe a index
print(df.loc[1])  # df.iloc[1] works same 
```

```python
print(df["calories"])
print("-"*20)
print(df["duration"])
```

```python
df.loc[[0,1]]  """Use to observe multiple index"""
```

```python
df = pd.DataFrame(data, index = ['day1', 'day2', 'day3'])
print(df)
```

```python
print(df.loc['day1'])
print('-'*20)
# print(df.loc[1])  once indics by string can't use integer as locator
print(df.iloc[1])  # can't use string index.
```
