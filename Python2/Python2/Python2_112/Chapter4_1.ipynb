### Data Preprocessing
1. Data Scaling = Standardize the range of features of data
2. Encoding Categorical variable(one hot encoding) --> numerical
                           --> categorical data --> to encode them to numerical values before fitting and evaluating model
3. Data splitting --> training, testing
4. Handeling the missing value
5. Features Selection --> reduce overfitting (make noice based decisions)

```python
import pandas as pd
df = pd.DataFrame({
    'OriginalFeature' : [1,2,3,4,5,6,7,8],
    'CategoricalFeature' : ['A','B','C', 'A', 'B', 'A', 'C','B']
})
df
```

```python
data = pd.get_dummies(df, columns = ['CategoricalFeature'])
data
```

```python
data = pd.get_dummies(df, columns = ['CategoricalFeature'], drop_first = True)
data
```

### Adding a new Feature

```python
import pandas as pd
room_len = [18, 20, 10, 12, 18, 11]
room_width = [20, 20, 10, 11, 19, 10]
room_type = ['Big', 'Big', 'Normal', 'Normal', 'Big', 'Normal']
data = pd.DataFrame({
    'Length' : room_len, 'width' : room_width, 'Type' : room_type
})
data
```

```python
data['Area'] = data['Length'] * data['width']
data
```
