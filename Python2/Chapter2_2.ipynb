## NetworkX

```python
import matplotlib.pyplot as plt
import networkx as nx

G = nx.Graph()
G.add_node(1)
G.add_nodes_from([2,3])
G.add_nodes_from(range(4,7))
G.add_edge(1,2)
G.add_edges_from([(2,3), (3,6), (4,6), (5,6)])
nx.draw_networkx(G, node_size = 850, node_color = 'red', width = 2, edge_color = 'blue')
plt.show()
```

```python
import matplotlib.pyplot as plt
import networkx as nx

G = nx.DiGraph()  # Directional Graph.
G.add_node(1)
G.add_nodes_from([2,3])
G.add_nodes_from(range(4,7))
G.add_edge(1,2)
G.add_edges_from([(2,3), (3,6), (4,6), (5,6)])
nx.draw_networkx(G, node_size = 850, node_color = 'red', width = 2, edge_color = 'blue')
plt.show()
```

```python
import matplotlib.pyplot as plt
import networkx as nx

G = nx.DiGraph()
l = [[0,1,0,1], [0,0,0,1], [1,0,0,0], [1,0,1,0]]
G.add_nodes_from([0,1,2,3])

for i in range(4):
    for j in range(4):
        if l[i][j] == 1:
            G.add_edge(i,j)
nx.draw(G, with_labels = True, node_size = 850,  node_color = 'red', width = 2, edge_color = 'blue')
plt.show()
```

```python
import matplotlib.pyplot as plt
import networkx as nx
l = {'A' : [1,0,0,1,0,1,1],
     'B' : [0,1,1,1,1,1,0],
     'C' : [1,1,1,0,1,0,1],
     'D' : [1,0,0,0,1,0,1],
     'E' : [1,0,0,0,1,1,0],
     'F' : [1,1,1,1,1,0,1],
     'G' : [0,1,1,0,0,0,1]}

G = nx.DiGraph()
key_list = list(l.keys())
G.add_nodes_from(key_list)
for key, value in l.items():
    for i in range(7):
        if value[i] == 1:
            G.add_edge(key, key_list[i])
nx.draw(G, with_labels = True, node_size = 850,  node_color = 'red', width = 2, edge_color = 'blue')
plt.show()  
```
